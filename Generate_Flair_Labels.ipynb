```python
from flair.models import TextClassifier
from flair.data import Sentence
import pandas as pd
```

```python
#Reading the dataset
dataset = pd.read_csv('clean_crawled_reviews_5k.csv')
dataset.drop(['Unnamed: 0'], axis=1, inplace=True)
dataset.head()
```

```python
len(dataset)
```

```python
classifier = TextClassifier.load('en-sentiment')
```

```python
labels_list = []
counter = 1

for row in dataset.itertuples():
  sentence = Sentence(row.Review)
  classifier.predict(sentence)

  if 'NEGATIVE' in str(sentence.labels[0]):
    labels_list.append(0)
  elif 'POSITIVE' in str(sentence.labels[0]):
    labels_list.append(1)
  else:
    print('ERROR: Not positive or negative, look through!')
    break  

  print(f"Row: {counter}")
  counter += 1
```

```python
print(f"Total Length of list: {len(labels_list)}")
print(f"Positives: {labels_list.count(1)}")
print(f"Negatives: {labels_list.count(0)}")
```

```python
dataset['flair_labels'] = labels_list
```
